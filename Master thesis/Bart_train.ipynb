```
from genre import GENRE

bart = (
    GENRE.from_pretrained("/content/drive/My Drive/seq2seq_coref/large_model_save", "checkpoint140.pt")
    .eval()
    .to("cpu:0")
)

codes = {
      k: [
          bart.encode(" {}".format(k))[1].item(),
      ]
      for k in ("{", "}", "[", "]")
  }
codes["EOS"] = [2]


def prefix_allowed_tokens_fn(batch_id, sent):
    sent = sent.tolist()
    status = get_status(sent)
    return get_allowed_tokens(
        sent, 
        sent_origs[batch_id], 
        status
    )

def apply_mask_fn(x, prev_output_tokens, original_batch_idxs):
    beam_size = x.shape[0] // original_batch_idxs.shape[0]
    original_batch_idxs = (
        original_batch_idxs.unsqueeze(-1)
        .repeat((1, beam_size))
        .flatten()
        .tolist()
    )

    mask = torch.full_like(x, -math.inf)
    for sent_i, (sent, batch_i) in enumerate(
        zip(prev_output_tokens, original_batch_idxs)
    ):
        sent = sent.tolist()
        status = get_status(sent)
        mask[
            sent_i, :, get_allowed_tokens(
                sent, 
                sent_origs[batch_i], 
                status
            )
        ] = 0
    return mask




def get_status(sent):
    c = [
        f
        for e in (
            codes["{"],
            codes["}"],
            codes["["],
            codes["]"],
        )
        for f in e
    ]
    status = (
        sum(
            e
            in c
            for e in sent
        )
        % 4
    )

    if status == 0:
        return "o"
    elif status == 1:
        return "m"
    else:
        return "e"

#prefix_allowed_tokens_fn(0, torch.tensor([2]))

# sent_origs = [[2] + model.encode(log).tolist()[1:]]

def get_allowed_tokens(sent, sent_orig, status):
    if status == "o":
        trie_out = get_trie_outside(sent, sent_orig)
    elif status == "m":
        trie_out = get_trie_mention(sent, sent_orig)
    elif status == "e":
        trie_out = get_trie_entity(sent, sent_orig)
        if trie_out == [2]:
            trie_out = get_trie_outside(sent, sent_orig)
    else:
        raise RuntimeError

    return trie_out

def get_pointer_end(sent, sent_orig):
    i = 0
    j = 0
    while i < len(sent):
        if sent[i] == sent_orig[j]:
            i += 1
            j += 1
        elif sent[i] in codes["{"] or sent[i] in codes["}"]:
            i += 1
        elif sent[i] in codes["["]:
            i += 1
            while sent[i] not in codes["]"]:
                i += 1
            i += 1
        else:
            return None

    return j if j != len(sent_orig) else None

def get_trie_outside(sent, sent_orig):
    pointer_end = get_pointer_end(sent, sent_orig)

    if pointer_end:
        if sent_orig[pointer_end] not in codes["EOS"]:
            return [sent_orig[pointer_end]] + codes["{"]
        else:
            return [sent_orig[pointer_end]]
    else:
        return []

def get_trie_mention(sent, sent_orig):
    pointer_end = get_pointer_end(sent, sent_orig)
    
    if pointer_end:
        if sent_orig[pointer_end] not in codes["EOS"]:
            return [sent_orig[pointer_end]] + codes["}"]
        else:
            return codes["}"]
    else:
        return []


def get_pointer_mention(sent):
    pointer_end = -1
    #pointer_start = 0
    for i, e in enumerate(sent):
        if e in codes["{"]:
            pointer_start = i
        elif e in codes["}"]:
            pointer_end = i

    return pointer_start, pointer_end

def get_trie_entity(sent, sent_orig):
    pointer_start, pointer_end = get_pointer_mention(sent)

    if pointer_start + 1 != pointer_end:
       # max_id = 5
        #candidates = [str(i) for i in range(1, max_id + 2)]
        candidates = get_candidates(sent)
        if candidates:
            return Trie([
                bart.encode(" }} [ {} ]".format(e)).tolist()[1:] 
                for e in candidates
            ]).get(sent[pointer_end:])
    
    #return []
    return pointer_start, pointer_end

def get_candidates(sent):
    counter = 0
    for i in range(len(sent)):
        if sent[i] in codes["["]:
            if i + 1 < len(sent):
                counter = max(int(bart.decode(torch.tensor([sent[i + 1]])).strip()), counter)
    return [str(e) for e in range(1, counter + 2)]

# model.sample(
#     ["London is the capital of the UK. London is also a city."],
#     prefix_allowed_tokens_fn=prefix_allowed_tokens_fn,
# )
```

```
from tqdm import tqdm
from collections import defaultdict
class Trie(object):
    def __init__(self, sequences):
        next_sets = defaultdict(list)
        for seq in sequences:
            if len(seq) > 0:
                next_sets[seq[0]].append(seq[1:])
        self._leaves = {k: Trie(v) for k, v in next_sets.items()}
    def get(self, indices):
        if len(indices) == 0:
            return list(self._leaves.keys())
        elif indices[0] not in self._leaves:
            return []

        else:
            return self._leaves[indices[0]].get(indices[1:])
```
